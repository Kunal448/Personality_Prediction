```python
def univariate_analysis(df):              #to make countplot for each columns
    for column in df.columns:
        if df[column].dtype == 'object':
            print("Univariate Analysis for:", column)
            print(df[column].value_counts())
            print("\n")

            # Plotting
            plt.figure(figsize=(5, 3))
            sns.countplot(data=df, x=column)
            plt.title("Count of " + column)
            plt.xticks(rotation=45)
            plt.show()
```

```python
def bivariate_analysis(df, target):       #to make countplot of columns and compare it with other columns
    for column in df.columns:
        if df[column].dtype == 'object' and column != target:
            print("Bivariate Analysis for:", column, "vs", target)
            plt.figure(figsize=(10, 6))
            sns.countplot(data=df, x=column, hue=target)
            plt.title("Count of " + column + " vs " + target)
            plt.xticks(rotation=45)
            plt.legend(title=target)
            plt.show()
```

```python
univariate_analysis(df)       #to call the function and display the graphs
```

```python
target = 'movie_preferences'
bivariate_analysis(df, target)       #here you can set target column and compare that with column using countplot
```

```python
def pie(df):         # this is to make pie chart for each column
    for column in df.columns:
        if df[column].dtype == 'object':
            value_counts = df[column].value_counts()

            # Plotting
            plt.figure(figsize=(5, 5))
            plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
            plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
            plt.title('Pie Chart of {}'.format(column))
            plt.show()
```

```python
pie(df)
```
